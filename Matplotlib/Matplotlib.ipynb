```python
x_axis = (1,2,3,4,5,6)
y_axis= (3,9,10,12,16,18)
# Display plot with x & y
a = plt.plot(x_axis,y_axis)
# plt.show()
```

```python
# Create 10 equal float numbers from 0 to 5 
x_1 = np.linspace(0,5,10)
y_1 = x_1**2
# Display plot with x & y 
# We can choose line style in plot function   
#ls is linestyle = '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
#lw is seekness of the line  
# alpha is for contrast degree for the chart
plt.plot(x_1, y_1,color = 'k',linestyle = 'dotted',linewidth = 8, alpha = 0.90)           
# Add  title for graph
plt.title('Days Squared Chart \n',color ='r')
# Add  X-axis Label 
plt.xlabel('Days',color = 'r')
# Add Y-axis Label
plt.ylabel('Days Squared',color = 'r')
plt.show()
```

```python
import datetime
goog_data = pd.read_csv('GOOG.csv')
#
goog_data_np = goog_data.to_numpy()
# Get array of prices in 5th column
goog_cp = goog_data_np[:,4]
goog_cp

# Get NumPy array with just weekdays between dates excluding holidays
holidays = [datetime.datetime(2020,5,25), datetime.datetime(2020,8,19)]
date_arr = pd.bdate_range(start='5/20/2020', end='8/19/2020', 
                          freq='C',
                         holidays=holidays)
date_arr_np = date_arr.to_numpy()
fig_7 = plt.figure(figsize=(8,5),dpi=100)
axes_7 = fig_7.add_axes([0.1,0.1,0.9,0.9])
plt.plot(date_arr_np, goog_cp)
plt.show()
```
